## Checkpoint 1: Data Cleaning 1

* Load the companies and rounds data into two data frames and name them companies and rounds2 respectively.

```python
# Code for importing the company file
companies  = pd.read_csv("companies.txt", sep="\t", encoding = "ISO-8859-1") 
companies.shape 
```

```python
#Code to remove special characters and encoding to UTF-8
companies['permalink'] =  companies['permalink'].str.encode('utf-8').str.decode('ascii', 'ignore')
companies
```

## Checkpoint 4: Sector Analysis 1

* Extract the primary sector of each category list from the category_list column

* Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also considered one of the main sectors)

* Expected Results: Code for a merged data frame with each primary sector mapped to its main sector (the primary sector should be present in a separate column).

```python
# Import mapping file to find main sectors
mapping  = pd.read_csv("mapping.csv", encoding = "ISO-8859-1") 
mapping.shape 
```

```python
mapping
```

```python
#removing blank column as it does not map to any existing primary sectors 
mapping = mapping.drop(['Blanks'], axis =1)
mapping.shape
```

```python
mapping
```

```python
#removing null value as it does not map to any existing primary sectors 
mapping = mapping[~mapping['category_list'].isnull()]
mapping.shape
```

```python
mapping
```

```python
#Cleaning out mapping dataframe as values containing 0 should be replaced by na. e.g. A0lytics as Analytics
mapping['category_list'] = mapping['category_list'].str.replace('0','na', regex=True)
mapping.shape
```

```python
mapping
```

```python
#creating primary sector column to contain only the first piped column
master_frame['primary_sector'] = master_frame['category_list'].str.split('|', n = 0, expand = True)[0]
master_frame
```

```python
#using melt function to unpivot the main sectors into long form. this would ease merge with parent df
melted = pd.melt(mapping, id_vars=["category_list"], var_name="main_sector", value_name="selection_index")

#pick only those primary sector - main sector combinations where its a match (1 means match)
melted = melted[melted['selection_index'] == 1]

#rename category_list to primary_sector for ease in merging with parent df. Would avoid duplication of key valuesin merged df
melted = melted.rename(columns={"category_list": "primary_sector"})

#this would be an unpivoted version of the mapping df
melted.shape
```

```python
melted
```

```python
#add main-sector values by mapping master_frame with melted df by merging on primary sector
master_frame = pd.merge(master_frame,melted[['primary_sector','main_sector']] , how = 'inner', on = 'primary_sector')
master_frame.shape
```

```python
master_frame
```
